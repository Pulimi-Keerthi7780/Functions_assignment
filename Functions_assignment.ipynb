10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- The map() function basically applies a function or an operation to all the elements in an iterable . map(function , iterable) is how you can use the map function.


```
# using map() to find the squares of elements in a list
l = [1,2,3]
k = list(map(lambda x : x**2 , l))
print(k)  #will print [1, 4, 9]
```


- The reduce() function basically reduces the entire items of an iterable into a single data depending on the function u give.


```
from functools import reduce
l = [47,11,42,13]
reduce(lambda x,y : x+y , l)   #will print 113
```


- The filter() function lets only the values that is True for a given condition be there in the given iterable


```
l = [1,2,3,4,5,6,6,7,88,98]
k = list(filter(lambda x : x%2==0 , l))
print(k)  # prints [2,4,6,6,88,98]
```

````python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list

def summ(liss):
  su = 0
  for i in liss:
    su = su + float(i)
  return su

mylist = input("Enter the list with elements seperated with commas : ")
lis = mylist.split(",")
listt = list(lis)
print(f"The sum of the elements in the given list is : {summ(listt)}")
````

````python
# 2.  Create a Python function that accepts a string and returns the reverse of that string.

def revstr(string):
  for i in range(len(string)):
    print(string[len(string)-i-1],end="")

s = input("Enter a string : ")
revstr(s)
````

````python
# 3.  Implement a Python function that takes a list of integers and returns a new list containing the squares of each number
def sqr(l):
  print([int(i)**2 for i in l])
my_list = input("Enter the numbers seperated by commas : ")
lis = my_list.split(",")
listt = list(lis)
sqr(listt)
````

````python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def prime(n):
  count = 0
  for i in range(2,(n//2)):
    if n%i == 0:
      count = count + 1
  if count >= 1:
    print("Not prime")
  else:
    print("Prime")

no = int(input("Enter a number : "))
if no==1 :
  print("Neither prime nor composite")
else:
  prime(no)
````

````python
# 5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms

n = int(input("Enter the number of terms required : "))
a = 0
b = 1
lst = [0]
for i in range(n-1):
  a,b = b , a+b
  i = i + 1
  lst.append(a)

x = iter(lst)
for i in range(n):
  print(next(x))
````

````python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent

def expo(n):
  yield 2**n

n = int(input("Enter the number : "))
for i in range(n+1):
  print(f"2 power {i} is : {next(expo(i))}")
````

````python
# 7.  Implement a generator function that reads a file line by line and yields each line as a string

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # yield each line without extra newline characters
file_path = 'file.txt'  # Replace with the path to your file

# Create the generator
line_generator = read_file_line_by_line(file_path)

# Iterate through the file line by line
for line in line_generator:
    print(line)  # Print each line from the file
````

````python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit
lst = input("Enter the temperatures seperated by commas : ")
ls = lst.split(",")
l = list(ls)
a = map(lambda x : (float(x)*(9/5) + 32) , l )
newlist = []
for i in range(len(l)):
  newlist.append(next(a))
print(newlist)
````

````python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string
def remove_vow(s):
  vow = "AEIOUaeiou"
  a = filter(lambda x : x not in vow , s)
  return "".join(a)

m = input("Enter a string : ")
print(remove_vow(m))

````

````python
# 11.  Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
'''
Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,-  if the value of the
order is smaller than 100,00 . Write a Python program using lambda and map.
'''

n = int(input("Enter the number of orders : "))
lst = []
for i in range(n):
  print(f"Please enter the details of {i+1} order :-")
  lst.append((input("Enter the order number : "),float(input("Enter the quantity : ")),float(input("Enter the price per item : "))))
#print(lst)

func = lambda x : (x[0],x[1]*x[2]) if x[1]*x[2] >= 100 else (x[0],x[1]*x[2]+10)
new = list(map(func , lst))

print(new)
````
